```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(1000)
```

```python
plt.hist(data,bins=30)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Freq")
plt.show()
```

```python
sizes=[30,25,15,10,5,5]
labels = ['A', 'B', 'C', 'D', 'E', 'F']

plt.pie(sizes, labels=labels, autopct='%1.0f%%')
plt.title("Pie Chart")
plt.show()
```
