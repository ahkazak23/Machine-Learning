```python
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.title("elbow method")
plt.xlabel("no of cluster")
plt.ylabel("wcss")
plt.show()
```
